### x axis lables are so close , these can be roatated using xticks() function and argument rotation

```python
X=unrate["DATE"][:12]
Y=unrate["UNRATE"][:12]
plt.plot(X,Y)
plt.xticks(rotation=90) #xticks() function can be used by argument rotation to raotate x axis labels.
```

### Plot can be more tweaked by adding axis labels and title.

### xlabel(): accepts a string value, which gets set as the x-axis label.

### ylabel(): accepts a string value, which is set as the y-axis label.


### title(): accepts a string value, which is set as the plot title.

```python
X=unrate["DATE"][:12]
Y=unrate["UNRATE"][:12]
plt.plot(X,Y)
plt.xticks(rotation=90) #xticks() function can be used by argument rotation to raotate x axis labels.
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
```

# STEP 9

Set the plotting area to a width of 10 inches and a height of 6 inches.
Generate the following plots in the base subplot:
1948: set the line color to "red"
1949: set the line color to "blue"
1950: set the line color to "green"
1951: set the line color to "orange"
1952: set the line color to "black"
Use plt.show() to display the plots.

```python
fig=plt.figure(figsize=(10,6)) #SETTING FIGURE SIZE TO 10,6

color={0:"red",1:"blue",2:"green",3:"orange",4:"black"} #SETTING COLORS TO PIC DYNAMIC

for i in range(5): 
    start_x=i*12
    end_x=(i+1)*12


    plt.plot(unrate['MONTH'][start_x:end_x],unrate['UNRATE'][start_x:end_x],c=color[i]) #ASSIGNING COLOR DYNAMICALLY
```

# STEP 10

Modify the code from the last screen that overlaid 5 plots to include a legend. <br>
Use the year value for each line chart as the label.<br>
E.g. the plot of 1948 data that uses "red" for the line color should be labeled "1948" in the legend.<br>
Place the legend in the "upper left" corner of the plot.<br>
Display the plot using plt.show().<br>

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black'] #SETTING COLORS TO PIC DYNAMIC
legend = [1948, 1949, 1950, 1951, 1952]    #SETTING LABEL TO PIC DYNAMIC
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i], label=legend[i]) #ASSIGNING LABEL TO PLOT DYNAMICALLY
    plt.legend(loc='upper left') #CREATING LEGEND IN UPPER LEFT TO PROVIDE MORE INFO ON EACH LINE
plt.show()
```

# STEP 11

Modify the code from the last screen:<br>
>-  Set the title to "Monthly Unemployment Trends, 1948-1952".<br>
-  Set the x-axis label to "Month, Integer".<br>
-  Set the y-axis label to "Unemployment Rate, Percent".


```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black'] #SETTING COLORS TO PIC DYNAMIC
legend = [1948, 1949, 1950, 1951, 1952]    #SETTING LABEL TO PIC DYNAMIC
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i], label=legend[i]) #ASSIGNING LABEL TO PLOT DYNAMICALLY
    plt.legend(loc='upper left') #CREATING LEGEND IN UPPER LEFT TO PROVIDE MORE INFO ON EACH LINE

plt.legend(loc='upper left')                       #ASSIGNING LEGEND TO PLOT
plt.title("Monthly Unemployment Trends, 1948-1952")#ASSIGNING TITLE TO PLOT
plt.xlabel("Month, Integer")                    #ASSIGNING X-AXIS LABEL
plt.ylabel("Unemployment Rate, Percent")        #ASSIGNING Y-AXIS LABEL

plt.show()
 
```
