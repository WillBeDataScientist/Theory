Modify the code from the last screen that overlaid 5 plots to include a legend. <br>

# 5. Aligning Axis Ticks And Labels

-  Create a single subplot and assign the returned Figure object to fig and the returned Axes object to ax.<br>
-  Generate a bar plot with:<br>
    -  left set to bar_positions<br>
    -  height set to bar_heights<br>
    -  width set to 0.5<br>
-  Set the x-axis tick positions to tick_positions.<br>
-  Set the x-axis tick labels to num_cols and rotate by 90 degrees.<br>
-  Set the x-axis label to "Rating Source".<br>
-  Set the y-axis label to "Average Rating".<br>
-  Set the plot title to "Average User Rating For Avengers: Age of Ultron (2015)".<br>
-  Use plt.show() to display the bar plot.<br>

```python
import matplotlib.pyplot as plt
from  numpy import arange

fig, ax=plt.subplots()
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
num_rows=['0%','20%','40%','60%','80%','100%']
bar_position=arange(5)
bar_heights=norm_reviews[num_cols].iloc[0].values
w=0.5 #width
tick_position=range(0,6) #manually feeding the ticks

ax.bar(bar_position,bar_heights,w)
ax.set_xticks(tick_position)
ax.set_xticklabels(num_cols,rotation=90)
ax.set_yticklabels(num_rows)
ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015).")
plt.plot()
```
