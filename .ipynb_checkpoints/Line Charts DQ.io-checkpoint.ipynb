# Line Charts

## import pandas library and read csv file by pd.read_csv() method

```python
import pandas as pd
unrate=pd.read_csv("UNRATE.csv")
```

## Print couple of data sets from read file

```python
print(unrate.head(3))
```

##### lets convert DATE which is in string to datetime format using pandas pd.to_datetime(column)

```python
unrate["DATE"]=pd.to_datetime(unrate["DATE"])
```

## import matplotlib.pyplot for ploting graphs 

```python
%matplotlib inline              
#need to use %matplotlib inline this line whenmatplot lib for each instance of jupyter notebook
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

#### above plot is default plot by matplotlib.pyplot with values -0.04 to 0.04

#let input some data and see line plot for 12 records.

```python
X=unrate["DATE"].head(12)
Y=unrate["UNRATE"].head(12)
plt.plot(X,Y)
```

### x axis lables are so close , these can be roatated using xticks() function and argument rotation

```python
X=unrate["DATE"][:12]
Y=unrate["UNRATE"][:12]
plt.plot(X,Y)
plt.xticks(rotation=90) #xticks() function can be used by argument rotation to raotate x axis labels.
```

### Plot can be more tweaked by adding axis labels and title.

### xlabel(): accepts a string value, which gets set as the x-axis label.

### ylabel(): accepts a string value, which is set as the y-axis label.


### title(): accepts a string value, which is set as the plot title.

```python
X=unrate["DATE"][:12]
Y=unrate["UNRATE"][:12]
plt.plot(X,Y)
plt.xticks(rotation=90) #xticks() function can be used by argument rotation to raotate x axis labels.
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
```

# Multiple Plots

```python
X=unrate["DATE"][:12]
Y=unrate["UNRATE"][:12]
plt.plot(X,Y)
plt.xticks(rotation=90) #xticks() function can be used by argument rotation to raotate x axis labels.
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
```

# step 5

Create 2 line subplots in a 2 row by 1 column layout:
In the top subplot, plot the data from 1948.
For the x-axis, use the first 12 values in the DATE column.
For the y-axis, use the first 12 values in the VALUE column.
In the bottom subplot, plot the data from 1949.
For the x-axis, use the values from index 12 to 24 in the DATE column.
For the y-axis, use the values from index 12 to 24 in the VALUE column.
Use plt.show() to display all the plots.

```python
fig=plt.figure()  #--------------> Giving Alias to Figure method             
ax1=fig.add_subplot(2,1,1) #-----> Saying what kind of plot we want, here 2 by 1 in 1 plot
ax2=fig.add_subplot(2,1,2) #-----> Saying what kind of plot we want, here 2 by 1 in 2 plot

x12=unrate["DATE"][:12] #--------> Collecting values for first plot x axis or first 12 
Y12=unrate["UNRATE"][:12]  #-----> Collecting values for first plot y axis or first 12
x1224=unrate["DATE"][12:24]  #---> Collecting values for second plot x axis or 12-24
Y1224=unrate["UNRATE"][12:24] #--> Collecting values for second plot y axis or 12-24

ax1.plot(x12,Y12)  #Plot on ax1 with x12,y12
ax2.plot(x1224,Y1224)#Plot on ax2 with x1224,y1224

plt.xlabel("Two Plots ")
plt.show()  #display Plot
```
